```python
#import the required libraries

import numpy as np # linear algebra
import pandas as pd 
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline

```

```python
#import sub features
from subprocess import check_output
from wordcloud import WordCloud, STOPWORDS
```

```python
#create the dataframe
df = pd.read_csv('data_project_final.csv', date_parser='name')
```

```python
#mpl.rcParams['figure.figsize']=(8.0,6.0)    #(6.0,4.0)
mpl.rcParams['font.size']=24              #10 
mpl.rcParams['savefig.dpi']=250           #72 
mpl.rcParams['figure.subplot.bottom']=.1 
```

```python
# Adding additional stopwords to the set

STOPWORDS.add('object')
STOPWORDS.add('Length')
STOPWORDS.add('dtype')
STOPWORDS.add('Name')

```

```python
#declaring the set stopwords
stopwords = set(STOPWORDS)
```

```python
#Generating the worldcloud with the website name
wordcloud = WordCloud(
                          background_color='white',
                          stopwords=stopwords,
                          max_words=100,
                          max_font_size=40, 
                          random_state=42
                         ).generate(str(df['name']))
```

```python
#Printing the wordcloud and storing in a png file
print(wordcloud)
fig = plt.figure(1)
plt.imshow(wordcloud)
plt.axis('off')
plt.show()
fig.savefig("wordcloud.png", dpi=1000)
```
